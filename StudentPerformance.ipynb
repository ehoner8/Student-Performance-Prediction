```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(pd.get_dummies(X, drop_first=True).astype(int), y["G3"], test_size=0.22, random_state=11)
```

We will begin with a linear regression model.

```python
X_train.columns
```

```python
X_train.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
```

```python
elastic_model = ElasticNet()
```

```python
param_grid = {"alpha": [0.01, 0.1, 0.5, 1, 5, 10, 20, 50, 70, 100, 150, 200], "l1_ratio": [.01, .1, .3, .5, .7, .9, .95, .99, 1]}
```

```python
grid_elastic_model = GridSearchCV(estimator=elastic_model, param_grid=param_grid, cv=10, verbose=1)
```

```python
grid_elastic_model.fit(scaled_X_train, y_train)
```

```python
grid_elastic_model.best_params_
```

```python
grid_elastic_preds = grid_elastic_model.predict(scaled_X_test)
```

```python
grid_elastic_preds
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, grid_elastic_preds)
```

```python
np.sqrt(mean_squared_error(y_test, grid_elastic_preds))
```

```python
from sklearn.svm import SVR
```

```python
svr_model = SVR()
```

```python
param_grid = {'C':[0.1,0.5,1,2,3,4,5,7,9],
             'kernel':['linear','rbf','poly'],
              'gamma':['scale','auto'],
              'degree':[2,3,4, 5, 6],
              'epsilon':[0,0.01,0.1,0.5,1,2]}
```

```python
grid_svr_model = GridSearchCV(estimator=svr_model, param_grid=param_grid, cv=10, verbose=1)
```

```python
grid_svr_model.fit(scaled_X_train, y_train)
```

```python
grid_svr_model.best_params_
```

```python
grid_svr_preds = grid_svr_model.predict(scaled_X_test)
```

```python
mean_absolute_error(y_test, grid_svr_preds)
```

```python
np.sqrt(mean_squared_error(y_test, grid_svr_preds))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor()
```

```python
len(X_train.columns)
```

```python
param_grid = {"n_estimators": np.arange(10, 300, 10), "max_features": np.arange(5, 27, 2), "bootstrap": [True, False]}
```

```python
grid_rf_model = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=8, verbose=1)
```

```python
grid_rf_model.fit(X_train, y_train)
```

```python
grid_rf_model.best_params_
```

```python
rmses = []
maes = []
for n in range(10, 300, 10):
    rfr = RandomForestRegressor(bootstrap=True, max_features=13, n_estimators=n)
    rfr.fit(scaled_X_train, y_train)
    y_preds = rfr.predict(scaled_X_test)
    rmses.append(np.sqrt(mean_squared_error(y_test, y_preds)))
    maes.append(mean_absolute_error(y_test, y_preds))
```

```python
plt.plot(range(10, 300, 10), rmses)
```

```python
plt.plot(range(10, 300, 10), maes)
```

```python
grid_rf_preds = grid_rf_model.predict(X_test)
```

```python
mean_absolute_error(y_test, grid_rf_preds)
```

```python
np.sqrt(mean_squared_error(y_test, grid_rf_preds))
```

```python
plt.figure(figsize=(8, 6), dpi=130)
sns.barplot(x=coefs.index, y=coefs.values, palette="Set2")
plt.xticks(rotation=90);
```

This shows which features are most important in predicting a student's final grade. The extreme positive features
in this graph indicate a strong predictor of getting a high final grade. The extreme negative features indicate the student
will get a low final grade.
